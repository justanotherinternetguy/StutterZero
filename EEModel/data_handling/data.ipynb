```python
import pandas as pd

# Load your CSV (no header, since it’s just rows of file paths)
df = pd.read_csv('/home/alien/Git/StutterZero-Git/EEModel/label,csv/filtered.csv', header=None)

# Function to extract the filename from the first column (or any of them)
def generate_test_path(row):
    # You can pick column 0 or 2 to extract filename, assuming they match
    filename = row[0].split('/')[-1]
    new_path = f"/home/alien/Git/StutterZero-Git/EEModel/test_wav/{filename}"
    return new_path


def generate_text_path(row):
    # You can pick column 0 or 2 to extract filename, assuming they match
    filename = row[0].split('/')[-1]
    filename = filename[:-4]
    new_path = f"/home/alien/Git/StutterZero-Git/EEModel/data_handling/whispertranscriptspredbetter/{filename}.txt"
    return new_path


# Apply the function and append as a new column (column 3)
df[3] = df.apply(generate_test_path, axis=1)
df[4] = df.apply(generate_text_path, axis=1)

# Save the updated CSV
df.to_csv('updated_file_2.csv', header=False, index=False)
```
